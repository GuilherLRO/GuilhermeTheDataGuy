```python
import requests

def enviar_para_ollama(prompt: str) -> str:
    """
    Envia um prompt para o Ollama local (modelo llama3.1:8b) e retorna a resposta.
    """
    url = "http://localhost:11434/api/generate"
    payload = {
        "model": "llama3.1:8b",
        "prompt": prompt,
        "stream": False
    }
    response = requests.post(url, json=payload)
    response.raise_for_status()
    resposta_json = response.json()
    # Alguns servidores Ollama retornam a resposta em 'response', outros em 'message'
    if 'response' in resposta_json:
        return resposta_json['response']
    elif 'message' in resposta_json and 'content' in resposta_json['message']:
        return resposta_json['message']['content']
    else:
        return resposta_json
```
